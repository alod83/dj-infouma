# Esercizio
Aprire il file `data/eurostat.csv` e caricarlo in un `DataFrame`

```python
import pandas as pd
df = pd.read_csv('data/eurostat.csv')
df
```

Selezionare solo i dati relativi all'Italia (colonna geo = IT)

```python
df = df[df['geo'] == 'IT']
df
```

Raggruppare i residenti per categoria `c_resid` (FOR, NAT, TOTAL) e per `unit` (NR, PCH_SM)  per ogni gruppo calcolare la somma per ogni anno

```python
groups = df.pivot_table(index=['c_resid','unit'])
groups
```

Per ogni categoria e per ogni unit, calcolare il valore medio su tutti gli anni. Usare la funzione `df.mean(axis=1)`. Calcolare anche il massimo e il minimo valore.

```python
groups.mean(axis=1)
```

```python
groups.max(axis=1)
```

```python
groups.min(axis=1)
```

Salvare il valore medio su file csv. Usare la funzione `to_csv()`.

```python
groups.mean(axis=1).to_csv("data/mean.csv")
```
